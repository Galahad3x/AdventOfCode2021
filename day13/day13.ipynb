```python
def fold_vertically(g, y):
    new_g = g.subgraph(vertices=[v for v in g.vertices() if v[0] < y])
    for vertex in g.vertices():
        if vertex[0] > y:
            new_0 = vertex[0] - (vertex[0] - y)*2
            if g.get_vertex(vertex) == '#' or new_g.get_vertex((new_0, vertex[1])) == '#':
                new_g.set_vertex((new_0, vertex[1]), '#')
            else:
                new_g.set_vertex((new_0, vertex[1]), '.')
    return new_g
def fold_horizontally(g, x):
    new_g = g.subgraph(vertices=[v for v in g.vertices() if v[1] < x])
    for vertex in g.vertices():
        if vertex[1] > x:
            new_1 = vertex[1] - (vertex[1] - x)*2
            if g.get_vertex(vertex) == '#' or new_g.get_vertex((vertex[0], new_1)) == '#':
                new_g.set_vertex((vertex[0], new_1), '#')
            else:
                new_g.set_vertex((vertex[0], new_1), '.')
    return new_g
def visible_dots(g):
    return list(g.get_vertices().values()).count('#')
```
